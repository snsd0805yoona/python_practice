```python
choice=input("What unit do you want to transform to? LB or KG: ")
unit=input("What unit do you enter? LB or KG: ")
userInput=int(input("Enter your weight: "))
```

```python
if(choice == "LB" and unit == "KG"):
    weight = userInput/0.45
    print(f"Your weight is {weight} LB")
elif(choice =="KG" and unit == "LB"):
    weight = userInput*0.45
    print(f"Your weight is {weight} KG")
else:
    print(f"Your weight is {userInput} {choice}")
```
