```python
import random
Secret_number=random.randrange(10)
guess_count = 0
guess_limit = 3
limit_count = 3
correct=False
while guess_count < guess_limit:
    userInput=int(input("Please enter your number: "))
    guess_count+=1
    if(userInput ==Secret_number):
        correct = True
        break;
    else:
        limit_count-=1
        print(f"You got the wrong number! Remain {limit_count} times! ")
        

if(correct == True):
    print("You get it! Congratulation!")
else:
    print("You didn't get it and your opportunity to guess the number is exhausted :(.")
```
