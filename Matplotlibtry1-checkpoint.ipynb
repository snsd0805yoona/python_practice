```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(25)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y, "b")

plt.subplot(1, 2, 2)
plt.plot(y,x, "r")
```

```python
fig = plt.figure(figsize=(15, 5))
axes1 = fig.add_subplot(1, 2, 1)
axes2 = fig.add_subplot(1, 2, 2)

axes1.plot(x, y, "b")
axes1.set_title("Title")
axes1.set_xlabel("X Axis", fontsize = 15)
axes1.set_ylabel("Y Axis", fontsize = 15)
axes1.plot(y, x, "r")
```
