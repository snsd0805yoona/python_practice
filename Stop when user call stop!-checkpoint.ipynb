```python
import datetime
command = ""
start = False
stop = False
while True:
    command=input("> ").lower()
    if(command == "start"):
        if(start):
            print("The car is already started.")
        else:
            start=True
            starttime=datetime.datetime.now()
            print(f"Car is starting at {starttime} !")
    elif(command == "stop"):
        if(stop):
            print("The car is already stopped.")
        else:
            stop=True
            stoptime=datetime.datetime.now()
            print(f"Car stopped at {stoptime} !")
    elif(command == "quit"):
        break
    elif(command == "help"):
        print("""
        Commands:
        start
        stop
        help
        quit
        
        """)
    else:
        print("I don't recognize your command! Please re-enter")
quittime=datetime.datetime.now()
print(f"User quit the car at {quittime} !")
```
